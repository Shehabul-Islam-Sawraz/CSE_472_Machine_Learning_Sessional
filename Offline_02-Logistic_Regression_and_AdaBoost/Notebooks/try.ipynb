### Telco Customer Churn Dataset Preprocessing

```python
# Import the required file
df = pd.read_csv("../Datasets/WA_Fn-UseC_-Telco-Customer-Churn.csv")

# To know the columns what are in this dataset
df.columns
```

```python
# Dropping unnecessary columns
df.drop(columns=['customerID'], inplace=True, axis=1)

df.columns
```

```python
# Top 5 rows in the dataset
df.head(5)
```

```python
# Show how many nulls in each column
print(df.isnull().sum())
print(any(df['TotalCharges'] == ' '))
```

```python
# Let's say TotalCharges column had null values, 
# we can fill this by placing ' ' in those null values 
df.TotalCharges = df.TotalCharges.fillna(' ')

print(df.isnull().sum())
```

```python
# Check the size of dataset and each column type

print(df.shape)
print("\n")
print(df.dtypes)
```

```python
# imputing missing values in specific columns
numeric_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

df['TotalCharges'].replace({' ': np.nan}, inplace=True)
df['TotalCharges'] = numeric_imputer.fit_transform(df[['TotalCharges']])

print(df['TotalCharges'])
```

```python
df = pd.get_dummies(
        df, 
        columns=[
            'gender', 'SeniorCitizen', 'Partner', 'Dependents', 
            'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
            'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 
            'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'
        ]
    )

df.columns
df.head(5)
```
